Задание 2.5

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
languages = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
popularity = [22.2,	17.6,	8.8,	8,	7.7,	6.7]
ax.bar(languages, popularity)

plt.minorticks_on()
plt.grid(which='major', linestyle='-', linewidth= '0.5', color='red')

plt.grid(which='minor', linestyle=':', linewidth= '0.5', color='black')

for language, pop in zip(languages, popularity):
    ax.text(language, pop + 0.3, f'{pop:.6f}', ha='center', va='bottom')

plt.legend()
plt.title("Popularity of Programming Language Worldwide, Oct 2017 compared to a year ago")
plt.xlabel("Languages")
plt.ylabel("Popularity")

plt.show()
```

1. Построить несколько графиков на одном поле

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 10))

sns.countplot(y='genre', data=df, ax=axes[0, 0])
axes[0, 0].set_title('Распределение по жанрам')

sns.scatterplot(x='chart', y='monthly_listens_total', data=df, ax=axes[0, 1])
axes[0, 1].set_title('Число прослушиваний в зависимости от чарта')

# График 3: Box plot для числа прослушиваний в зависимости от является ли контент эксплицитным
sns.boxplot(x='Explicit_content', y='monthly_listens_total', data=df, ax=axes[1, 0])
axes[1, 0].set_title('Число прослушиваний в зависимости от Explicit контента')

top_artists = df.groupby('artist(s)').mean().sort_values(by='monthly_listens_total', ascending=False).head(10)
sns.barplot(y=top_artists.index, x='monthly_listens_total', data=top_artists, ax=axes[1, 1])

# Подстройка макета
plt.tight_layout()

# Показ графиков
plt.show()
```

**2. Построить диаграммы для категориальных данных**

```python
plt.figure(figsize=(10, 8))
top_artists = df.groupby('artist(s)').mean().sort_values(by='monthly_listens_total', ascending=False).head(10)
sns.barplot(y=top_artists.index, x='monthly_listens_total', data=top_artists)
plt.title('Heatmap для частоты встречаемости пар категорий')
plt.show()
```

```python
plt.figure(figsize=(10, 8))
crosstab = pd.crosstab(df['genre'], df['Explicit_content'])
sns.heatmap(crosstab, annot=True, cmap='Blues')
plt.title('Heatmap для частоты встречаемости пар категорий')
plt.show()
```

**3. Построить столбчатые и круговые диаграммы**

```python
genre_counts = df['genre'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(genre_counts, labels=genre_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Доля каждого жанра')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='genre', y='monthly_listens_total', data=df)
plt.title('Monthly Listens Total по жанрам')
plt.xlabel('Жанр')
plt.ylabel('Monthly Listens Total')
plt.xticks(rotation=45)
plt.show()
```

```python
numeric_columns = df.select_dtypes(include=['int64']).columns

# Создание корреляционной матрицы
correlation_matrix = df[numeric_columns].corr()

# Построение heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=.5)
plt.title('Heatmap корреляции числовых признаков')
plt.show()
```

```python
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Данные для осей X, Y и Z
x = df['chart']
y = df['monthly_listens_total']
z = df['artists_likes_total']

# Построение 3D-графика
ax.scatter(x, y, z, c='blue', marker='o')

# Настройка подписей осей
ax.set_xlabel('Chart Position')
ax.set_ylabel('Monthly Listens Total')
ax.set_zlabel('Artists Likes Total')

# Настройка заголовка графика
plt.title('3D-график данных')

# Отображение графика
plt.show()
```
